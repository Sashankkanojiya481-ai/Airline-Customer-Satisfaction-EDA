```python
import os

print(os.listdir(r"C:\Users\Hp\Downloads\archive"))
import pandas as pd

df = pd.read_excel(
    r"C:\Users\Hp\Downloads\archive\satisfaction.xlsx"
)

df.head()
df.shape
df.info()
df.describe()
df.isnull().sum() 
df = df.dropna()

print(df.shape)
print(df.isnull().sum())
print(df["satisfaction_v2"].value_counts())
print(
    df["satisfaction_v2"]
    .value_counts(normalize=True) * 100
)
pd.crosstab(
    df["Class"],
    df["satisfaction_v2"]
)
pd.crosstab(
    df["Class"],
    df["satisfaction_v2"],
    normalize="index"
)*100

import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(
    x="Customer Type",
    hue="satisfaction_v2",
    data=df
)

plt.title("Customer Type vs Satisfaction")
plt.show()

df.groupby(
    "satisfaction_v2"
)[
    "Arrival Delay in Minutes"
].mean()

df.groupby(
    "satisfaction_v2"
)[
    "Departure Delay in Minutes"
].mean()

import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(
    x="satisfaction_v2",
    y="Arrival Delay in Minutes",
    data=df
)

plt.title("Arrival Delay vs Passenger Satisfaction")
plt.show()

df.groupby("satisfaction_v2")[
    ["Seat comfort", "Online boarding", "Inflight entertainment"]
].mean()

df.groupby("satisfaction_v2")["Age"].mean()

import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(
    x="satisfaction_v2",
    y="Age",
    data=df
)

plt.title("Age vs Passenger Satisfaction")
plt.show()

plt.savefig("satisfaction_distribution.png")
```
